```python
import pickle

import numpy as np

dataset_path = "../data/match_dataset"
train = pickle.load(open(dataset_path + "/train.pkl", "rb"))
```

```python
test = pickle.load(open(dataset_path + "/test.pkl", "rb"))
test[:10]
```

```python
data = train + test
len(data)
```

```python
train, val, test = (
    data[: int(0.7 * len(data))],
    data[int(0.7 * len(data)): int(0.75 * len(data))],
    data[int(0.75 * len(data)):],
)
```

```python
import json
import pickle
from tqdm import trange


def save_data(data, mode="train"):
    file_list = []
    for i in trange(len(data)):
        part_data = data[i]
        fname = f"{mode}/{mode}_part{i + 1}.pkl"
        file_list.append(fname)
        with open(f"{dataset_path}/{fname}", "wb") as f:
            pickle.dump(part_data, f)
    print("Data saved to", f"{dataset_path}/{mode}")
    with open(f"{dataset_path}/{mode}_parts.json", "w") as f:
        json.dump(file_list, f)


save_data(val, mode="validate")
```

```python
save_data(test, mode="test")
```

```python
save_data(train, mode="train")
```
