```python
import numpy as np  # numpy backend
import pygmtools as pygm
import matplotlib.pyplot as plt  # for plotting
from matplotlib.patches import ConnectionPatch  # for plotting matching result
import networkx as nx  # for plotting graphs

pygm.set_backend("numpy")  # set default backend for pygmtools
np.random.seed(1)  # fix random seed
```

```python
def create_random_graph(n, m):
    # 创建一个n x n的零矩阵
    graph = np.zeros((n, n))

    edges=0
    # 添加m条边
    while edges <= m+1: 
        # 随机选择两个节点
        i, j = np.random.choice(n, 2)
        w=np.random.rand()
        if i == j or w==0:
            continue
        # 添加边
        graph[i, j] = graph[j, i] = w
        edges+=1

    return graph


# 创建一个具有5个节点，7条边的随机图
graph = create_random_graph(3, 3)
graph
```

```python
nx.draw(nx.from_numpy_array(graph), with_labels=True)
```

```python
n1,n2=4,5
g1=create_random_graph(n1, n1)
g2=create_random_graph(n2, n2)
g1[:3,:3]=graph
g2[:3,:3]=graph
g1,g2
```
